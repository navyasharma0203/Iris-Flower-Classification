# **Iris Flower Classification using ML**

1. Import necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

2. Load the dataset

```python
# Load the data
df = pd.read_csv('IRIS.csv')
columns = ['Sepal_length', 'Sepal width', 'Petal length', 'Petal width', 'Class_labels']
df.columns = columns
df.head()
```

3. Analysing the dataset

```python
df.describe()
```

4. Visualizing the dataset

```python
sns.pairplot(df, hue='Class_labels')
plt.show()
```

5. Separating Input and Output Columns

```python
# Seperate features and target
data = df.values
X = data[:, 0:4]
Y = data[:, 4]
print (X)
```

6. Splitting the data into training and testing sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

MODEL 1 : Support Vector machine algorithm

```python
#SVM
from sklearn.svm import SVC
model_svc = SVC()
model_svc.fit(X_train, y_train)
```

```python
prediction1 = model_svc.predict(X_test)
#Determining accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction1)*100)
```

MODEL 2: Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model_LR = LogisticRegression()
model_LR.fit(X_train, y_train)
```

```python
prediction2 = model_LR.predict(X_test)
#Determining the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction2)*100)
```

MODEL 3: Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
model_DTC = DecisionTreeClassifier()
model_DTC.fit(X_train, y_train)
```
